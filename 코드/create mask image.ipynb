```python
import numpy as np
import os
os.mkdir("/content/dataset")
os.mkdir("/content/dataset/trainA")
os.mkdir("/content/dataset/testA")
os.mkdir("/content/dataset/trainB")
os.mkdir("/content/dataset/testB")

healthy = "/content/콩/A"
disease = "/content/콩/C"

len_all_A = len(os.listdir(healthy))
len_all_B = len(os.listdir(disease))

listA = np.array(os.listdir(healthy))
listB = np.array(os.listdir(disease))

from sklearn.model_selection import train_test_split
train_A_i, test_A_i, _, _ = train_test_split(list(range(len_all_A)), list(range(len_all_A)), random_state=0)
train_B_i, test_B_i, _, _ = train_test_split(list(range(len_all_B)), list(range(len_all_B)), random_state=0)

import shutil
for i in listA[train_A_i]:
    src = os.path.join(healthy, i)
    dst = os.path.join("/content/dataset/trainA", i)

    shutil.move(src, dst)

for i in listA[test_A_i]:
    src = os.path.join(healthy, i)
    dst = os.path.join("/content/dataset/testA", i)

    shutil.move(src, dst)

for i in listB[train_B_i]:
    # src = os.path.join("/content/콩/B", i)
    src = os.path.join(disease, i)
    dst = os.path.join("/content/dataset/trainB", i)

    shutil.move(src, dst)

for i in listB[test_B_i]:
    # src = os.path.join("/content/콩/B", i)
    src = os.path.join(disease, i)
    dst = os.path.join("/content/dataset/testB", i)

    shutil.move(src, dst)
```
